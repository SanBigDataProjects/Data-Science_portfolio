# declare x nad y 


```python
x = data.iloc[:, :-1]  # Select all rows, all columns except the last one (features)
y = data.iloc[:, -1]   # Select all rows, just the last column (target variable)
x
```

# spiliting tarining data and testing data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
 from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
x_test_scaled
```

```python
cols= x_train.columns
```

# Run SVM with default hyperparameters 

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
cols = x_train.columns
x_train_scaled = pd.DataFrame(x_train_scaled, columns=cols)
x_test_scaled = pd.DataFrame(x_test_scaled, columns=cols)

# Train the SVM model
svc = SVC()
svc.fit(x_train_scaled, y_train)

# Make predictions
y_predict = svc.predict(x_test_scaled)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_predict)
print("Accuracy:", accuracy)
```

# Run SVM with rbf kernel and C=100.0¶
We have seen that there are outliers in our dataset. So, we should increase the value of C as higher C means fewer outliers. So, I will run SVM with kernel=rbf and C=100.0.

```python
svc1=SVC(C=100.0)
svc1.fit(x_train_scaled,y_train)
y_predict=svc1.predict(x_test_scaled)
accuracy=accuracy_score(y_test,y_predict)
accuracy
```

# here when i incraese c value my accuracy got decrese

```python
svc1=SVC(C=1000.0)
svc1.fit(x_train_scaled,y_train)
y_predict=svc1.predict(x_test_scaled)
accuracy=accuracy_score(y_test,y_predict)
accuracy
```

#  Run SVM with linear kernel

```python
linear_svc=SVC(kernel='linear',C=1.0)
linear_svc.fit(x_train_scaled,y_train)
y_predict=linear_svc.predict(x_test_scaled)

accuracy=accuracy_score(y_test,y_predict)
accuracy
```

# Run SVM with linear kernel and C=100.0

```python
linear_svc=SVC(kernel='linear',C=100.0)
linear_svc.fit(x_train_scaled,y_train)
y_predict=linear_svc.predict(x_test_scaled)

accuracy=accuracy_score(y_test,y_predict)
accuracy
```

We can see that we can obtain higher accuracy with C=100.0 and C=1000.0 as compared to C=1.0.
