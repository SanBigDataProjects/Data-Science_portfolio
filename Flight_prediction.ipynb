```python
def flight_dep_time(x):
    if x <= 4 and x<=8:
        return "Early Morning"
    elif x >8 and x <=12:
        return "Morning"
    elif x >12 and x<=16:
        return "Noon"
    elif x >16 and x <=20:
        return "Evening"
    elif x > 20 and x <=24:
        return "Night"
    else: return "Late_Night"
```

```python
def preprocess_duration(x):
    if 'h' not in x:
        x = '0h ' + x  # Add '0h' if hours are missing
    elif 'm' not in x:
        x = x + ' 0m'  # Add '0m' if minutes are missing
    return x
```

```python
def plot(data, col):
    # Create 3 subplots
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 15))
    
    # Distribution plot (using histplot)
    sns.histplot(data[col], ax=ax1, kde=True)
    ax1.set_title(f'Distribution Plot of {col}')
    
    # Box plot
    sns.boxplot(data=data, x=col, ax=ax2)
    ax2.set_title(f'Box Plot of {col}')
    
    # Histogram without KDE
    sns.histplot(data[col], ax=ax3, kde=False)
    ax3.set_title(f'Histogram of {col}')

    plt.tight_layout()
    plt.show()
```

```python
def detect_outliers(data, columns):
    outliers = {}
    for col in columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Find outliers
        outliers[col] = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
    
    return outliers

# Detect outliers in numeric columns


num_col = ['Price', 'Duration_total_min', 'Dep_hour', 'Arrival_hour', 'Duration_hour', 'Duration_min']
outliers = detect_outliers(data_encoded, num_col)
```

```python
for col, outlier_data in outliers.items():
    print(f"Outliers in {col}:")
    print(outlier_data)
    print("\n")
```

```python
# Function to remove outliers
def remove_outliers(data, columns):
    for col in columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Remove outliers
        data = data[(data[col] >= lower_bound) & (data[col] <= upper_bound)]
    
    return data

# Remove outliers from data
cleaned_data = remove_outliers(data_encoded, num_col)
```

```python
# After removing or capping outliers, you can plot boxplots again
import matplotlib.pyplot as plt
import seaborn as sns

# Plot boxplots after handling outliers
def plot_boxplots(data, columns):
    fig, axes = plt.subplots(2, 9, figsize=(18, 8))
    axes = axes.flatten()
    for i, col in enumerate(columns):
        sns.boxplot(data=data, x=col, ax=axes[i])
        axes[i].set_title(f'Boxplot of {col}')
    plt.tight_layout()
    plt.show()


num_col= [col for col in data_encoded.columns if data_encoded[col].dtypes!="object"]

# Visualize after removing outliers
plot_boxplots(cleaned_data, num_col)

```

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

# Select only numeric columns
numeric_data = data_encoded[["Total_Stops", "Airline", "Duration_hour", "Duration_total_min", "Destination", "Dep_min", "Arrival_min"]]

# Add a constant to the data (for the intercept term in VIF calculation)
numeric_data = add_constant(numeric_data)

# Calculate VIF for each feature
vif_data = pd.DataFrame()
vif_data["Feature"] = numeric_data.columns
vif_data["VIF"] = [variance_inflation_factor(numeric_data.values, i) for i in range(len(numeric_data.columns))]

# Display the VIF for each feature
print(vif_data)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Select the features and target
x = data_encoded[["Total_Stops", "Airline", "Destination", "Dep_min", "Arrival_min"]]
y = data_encoded["Price"]

# Split the data into training and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Scale the features using StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)  # Fit and transform the training data
x_test_scaled = scaler.transform(x_test)  # Only transform the test data

# Check the shapes of the data
print(x_train_scaled.shape, y_train.shape, x_test_scaled.shape, y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor


# Fit the Random Forest model
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(x_train_scaled, y_train)  # Fit the model to the training data
```

```python
y_pred= model.predict(x_test_scaled)
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R^2 Score:", r2)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Define models to test
models = {
    'Random Forest': RandomForestRegressor(random_state=42),
    'Linear Regression': LinearRegression(),
    'Decision Tree': DecisionTreeRegressor(random_state=42),
    'Gradient Boosting': GradientBoostingRegressor(random_state=42),
    'K-Nearest Neighbors': KNeighborsRegressor()
}

# Store results
results = {}

# Loop through models
for name, model in models.items():
    # Train the model
    model.fit(x_train_scaled, y_train)
    
    # Make predictions
    y_pred = model.predict(x_test_scaled)
    
    # Calculate MSE and R²
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    # Store results
    results[name] = {'MSE': mse, 'R²': r2}

# Display the results
results_df = pd.DataFrame(results).T
print(results_df)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

# Define the parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}

# Instantiate the model
rf = RandomForestRegressor(random_state=42)
gb = GradientBoostingRegressor(random_state=42)

# Setup the GridSearchCV
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
rf_cv_scores = cross_val_score(rf, x_train_scaled, y_train, cv=5, scoring='neg_mean_squared_error')
gb_cv_scores = cross_val_score(gb, x_train_scaled, y_train, cv=5, scoring='neg_mean_squared_error')


# Fit the model
grid_search.fit(x_train_scaled, y_train)

# Best parameters and best score
print("Best Parameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)

# Evaluate on the test set
best_model = grid_search.best_estimator_
y_pred = best_model.predict(x_test_scaled)

# Calculate and print MSE and R^2 score
from sklearn.metrics import mean_squared_error, r2_score
print("MSE after tuning:", mean_squared_error(y_test, y_pred))
print("R^2 after tuning:", r2_score(y_test, y_pred))
```
